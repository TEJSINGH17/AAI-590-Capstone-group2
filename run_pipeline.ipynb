```python
import os
test_data = "/content/AAI-590-Capstone-group2/test_data"
if os.path.exists(test_data):
    files = os.listdir(test_data)
    print(f"Files in test_data/: {files}")
    for f in files:
        size = os.path.getsize(os.path.join(test_data, f))
        print(f"  {f}: {size/1e6:.1f} MB")
else:
    print("test_data/ folder does not exist!")
```
